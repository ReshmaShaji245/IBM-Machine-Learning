```python
import pandas as pd
import numpy as np
```

```python
concrete_data=pd.read_csv( 'https://cocl.us/concrete_data')
```

```python
concrete_data.head()
```

```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

```python
predictors.head()
```

```python
target.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
from sklearn.metrics import mean_squared_error
```

PART A

```python
mse=[]
for i in range(50):
    x_train,x_test,y_train,y_test=train_test_split(predictors,target,test_size=0.3)
    n_cols = x_train.shape[1]
# define regression model
    def regression_model():
    # create model
        model = Sequential()
        model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    
        model.add(Dense(1))
    
    # compile model
        model.compile(optimizer='adam', loss='mean_squared_error')
        return model
    model=regression_model()
    model.fit(x_train, y_train, validation_split=0.3, epochs=50, verbose=2)
    y_hat=model.predict(x_test)
    mse.append(mean_squared_error(y_test,y_hat))
print(mse)
```

```python
import statistics
print("Standard Deviation is % s" %(statistics.stdev(mse)))
mean=sum(mse)/len(mse)
print("Mean is %s"%mean)
```

PART B

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
mse_norm=[]
for i in range(50):
    x_train,x_test,y_train,y_test=train_test_split(predictors_norm,target,test_size=0.3)
    n_cols = x_train.shape[1]
# define regression model
    def regression_model_norm():
    # create model
        model_norm = Sequential()
        model_norm.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    
        model_norm.add(Dense(1))
    
    # compile model
        model_norm.compile(optimizer='adam', loss='mean_squared_error')
        return model_norm
    model_norm=regression_model_norm()
    model_norm.fit(x_train, y_train, validation_split=0.3, epochs=50, verbose=2)
    y_hat=model_norm.predict(x_test)
    mse_norm.append(mean_squared_error(y_test,y_hat))
print(mse_norm)
```

```python
import statistics
print("Standard Deviation is % s" %(statistics.stdev(mse_norm)))
mean=sum(mse_norm)/len(mse_norm)
print("Mean is %s"%mean)
```
